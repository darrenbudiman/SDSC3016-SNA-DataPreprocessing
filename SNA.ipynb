# create data set for machine learning

```python
cc = pd.read_excel('city-city edges.xlsx')
cc
```

```python
# get all unique cities
cities = set(cc['Source'].unique()).union(set(cc['Target'].unique()))

# create a new dataframe with all possible city connections
new_df = pd.DataFrame(columns=cities, index=cities)

# fill the new dataframe with 0s and 1s
for index, row in cc.iterrows():
    new_df.loc[row['Source'], row['Target']] = 1

new_df.fillna(0, inplace=True)

new_df
```

```python
# get all unique cities
cities = set(cc['Source'].unique()).union(set(cc['Target'].unique()))

# create a list of tuples with all possible city connections
city_connections = [(i, j) for i in cities for j in cities]

# create a new dataframe with columns "first city name", "second city name", and "connection"
new_df = pd.DataFrame(columns=['first city name', 'second city name', 'connection'])

# fill the new dataframe with 0s and 1s
for i, j in city_connections:
    if (i,j) in zip(cc['Source'], cc['Target']):
        new_df = new_df.append({'first city name': i, 'second city name': j, 'connection': 1}, ignore_index=True)
    else:
        new_df = new_df.append({'first city name': i, 'second city name': j, 'connection': 0}, ignore_index=True)
        
new_df
```

```python
# Create a multi-index with all unique cities
cities = cc['Source'].unique()
mi = pd.MultiIndex.from_product([cities, cities], names=['first_city', 'second_city'])

# Create a new empty dataframe for the connections
connections = pd.DataFrame({'link': 0}, index=mi)

# Set the connection value to 1 for all existing connections
cc['link'] = 1
connections.update(cc.set_index(['Source', 'Target'])['link'])

# Reset the index and rename the columns
connections = connections.reset_index().rename(columns={'first_city': 'First City', 'second_city': 'Second City'})

# Print the connections dataframe
connections
```

```python
# safe to excel for machine learning
connections.to_excel('connections.xlsx')
```
